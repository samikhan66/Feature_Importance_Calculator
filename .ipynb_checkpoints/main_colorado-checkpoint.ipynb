```python
# Problem: Build a model that explains how some of the indicators influence energy burden.
# For this notebook, I have only included the rows for the state of Colorado.
import os
import random

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, cross_val_score, RepeatedKFold
from sklearn.tree import DecisionTreeRegressor
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import RFECV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error
from xgboost import XGBRegressor

import tensorflow as tf
from tensorflow.keras.models import *
from tensorflow.keras.layers import *
```

```python
missing_values = ["n/a", "na", "--"]
# Read csv
df = pd.read_csv("interview_sample_data.csv", na_values=missing_values)
# Include rows for the state of Colorado only
df = df[df.state == "CO"]
# Find the statistical attributes of the dataframe
df.describe()
```

```python
# Find null values 
df.isnull().sum()
# Result shows that 85% of asthma_rate are Nans
```

```python
# Visualize the attributes to get a sense of the data
df.hist(bins=50, figsize=(20, 15))
plt.show()

# The histograms portray some of the standardizations that need to take place:
# 1. The indicators need to be converted to the same units (decimals).
# 2. The negative values in average cost columns need to be replaced with NaN . 
```

```python
# Replace negative numbers with NaN
avg_values_list = ["avg_cost_electricity", "avg_cost_gas", "avg_cost_water"]
df[avg_values_list] = df[avg_values_list].mask(df[avg_values_list] < 0)
```

```python
# Drop rows with na values since not too many. 
df = df.dropna()
print(df)
```

```python
# Convert eviction, poverty rate to decimal
df[["eviction_rate", "poverty_rate"]] = df[["eviction_rate", "poverty_rate"]] / 100

# Create energy burden column i.e. (avg cost of electricity + avg cost of gas) / median income
df["energy_burden"] = (df.loc[:, [avg_values_list[0], avg_values_list[1]]].sum(axis=1)) / df["median_income"]

# Replace median income with log(median income) to break linearity b/w income and energy burden
df["median_income"] = np.log10(df["median_income"])
```

```python
# Heatmap to visualize the correlated and uncorrelated data
corr_ = df.corr()["energy_burden"].sort_values(ascending = False)
corr_ = df.corr()
# visualise the data with seaborn
mask = np.triu(np.ones_like(corr_, dtype=bool))
sns.set_style(style='white')
f, ax = plt.subplots(figsize=(12, 15))
cmap = sns.diverging_palette(10, 250, as_cmap=True)
sns.heatmap(corr_, mask=mask, cmap=cmap,
            square=True, annot=True,
            linewidths=.5, cbar_kws={"shrink": .5}, ax=ax)
plt.show()
```

```python
# Use Recursive Feature Elimination to filter out features that are closely correlated and features 
# that have very low correlation with the target variable

# Define X(exogenous) and y(endogenous) variables that will be fed into the recursive elimination algorithm.
X = df.drop(['energy_burden'], axis=1)
y = df['energy_burden']
# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Use automatic feature selection instead of manually specifying number of features.
rfecv = RFECV(estimator=DecisionTreeRegressor())
model = DecisionTreeRegressor()
# Pipeline  since well perform some cross-validation. Its best practice in order to avoid data leakage.
pipeline = Pipeline(steps=[('Feature Selection',rfecv),('Model',model)])
# Evaluate model
cv = RepeatedKFold(n_splits=10, n_repeats=5, random_state=42)
n_scores = cross_val_score(pipeline, X_train, y_train, scoring='neg_mean_absolute_error', cv=cv, n_jobs=-1, error_score='raise')
# Report performance
print('MAE: %.3f (%.3f)' % (np.mean(n_scores), np.std(n_scores)))
pipeline.fit(X_train, y_train)
print("Optimal number of features : %d" % rfecv.n_features_)
print('\n')
# Find the 'Ranks' of each feature and sort these values by individual 'Rank'.
# Rank is the order of importance i.e. Rank 1 (most important), Rank 2 (second order of importance) and so on.
rf_df = pd.DataFrame(rfecv.ranking_,index=X.columns,columns=['Rank']).sort_values(by='Rank',ascending=True)
print(rf_df.head(rfecv.n_features_))
```

```python
# Recursive feature elimination resulted that the optimal number of features is five.
X = df.drop(['percent_renter', 'percent_asian', 'percent_hispanic', 'percent_multifamily_housing',
             'eviction_rate', 'avg_cost_water', 'energy_burden'], axis=1)
y = df['energy_burden']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Split dataset to train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
#Find existing columns
cols = [col for col in X.columns]
print(cols)
```

```python
def importance_fn(model):
    """
    Parameters
    ----------
        model : Model that has been fitted on the training data.

    Returns
    -------
        feature_importance: Calculate importance scores of each feature that
        the energy burden depends on.
    """
    # get importance
    importance = model.feature_importances_
    # summarize feature importance
    feature_importance = [value for value in importance]

    return feature_importance
```

```python
def random_forest():    
    """
    Find the feature importance that random forest predicts.
    Decision Trees are prone to overfitting so Random Forest algorithm was implemented.
    RF has enough capacity to completely memorize the training set
    but it can still generalize well enough due to built-in bagging of random forests.

    Returns
    -------
        rf_feature_importance : Calculate importance scores of each feature that
        the energy burden depends on.
        rf_train_accuracy : Accuracy of the Random Forest model on training data.
        rf_test_accuracy : Accuracy of the Random Forest model on test data.
    """

    model = RandomForestRegressor()
    # fit the model
    model.fit(X_train, y_train)
    rf_train_accuracy = model.score(X_train, y_train)
    rf_test_accuracy = model.score(X_test, y_test)
    rf_feature_importance = importance_fn(model)

    return importance_fn(model), rf_train_accuracy, rf_test_accuracy
```

```python
        
def plot_features(model=""):    
    """
    Plot feature importances predicted by selected algorithm while plotting it against features.
    """
    try:
        if model == "random_forest":
            rf_feature_importance, _, _ = random_forest()
            # plot feature importance for random forest
            plt.figure(figsize=(20, 8))
            plt.title("Feature Importance for Random Forest")
            plt.xlabel("Features")
            plt.ylabel("Importance")
            plt.bar([col for col in cols], [x for x in rf_feature_importance])
            plt.show()
        elif model == "xgboost":
            xgboost_feature_importance, _, _ = xg_boost()
            # plot feature importance
            plt.figure(figsize=(20, 8))
            plt.title("Feature Importance for XGBoost")
            plt.xlabel("Features")
            plt.ylabel("Importance")
            plt.bar([col for col in cols], [x for x in xgboost_feature_importance])
            plt.show()
        elif model == "dnn":
            # plot feature importance
            dnn_feature_importance = dnn()
            plt.figure(figsize=(20, 8))
            plt.title("Feature Importance for Deep Neural Network")
            plt.xlabel("Features")
            plt.ylabel("Importance")
            plt.bar([col for col in cols], [x for x in dnn_feature_importance])
            plt.show()
        elif model == "average":
            rf_feature_importance, _, _ = random_forest()
            xgboost_feature_importance, _, _ = xg_boost()
            dnn_feature_importance = dnn()
            avg_importance = [np.mean(i) for i in zip(rf_feature_importance, 
                                                      xgboost_feature_importance, 
                                                      dnn_feature_importance)]
            plt.figure(figsize=(20, 8))
            plt.title("Average Feature Importance of all models")
            plt.xlabel("Features")
            plt.ylabel("Importance")
            plt.bar([col for col in cols], [x for x in avg_importance])

            plt.show()
        else:
            pass
    except Exception as e:
        print(e)
plot_features()
```

```python
# Plot feature importances predicted by Random Forest algorithm while plotting it against features.
plot_features("random_forest") 
```

```python
def xg_boost():    
    """
    XGBoost for feature importance on a regression problem.

    Just like gradient boosting, XGBoost comprises an ensemble method
    XGBoost comes with an additional custom regularization term in the
    objective function which is why it outperforms other implementations
    of gradient tree boosting.

    Returns
    -------
        xgboost_feature_importance : Calculate importance scores of each feature that
        the energy burden depends on.
        xgboost_train_accuracy : Accuracy of the XGBoost model on training data.
        xgboost_test_accuracy : Accuracy of the XGBoost model on test data.
    """
    # define the model
    model = XGBRegressor()
    # fit the model
    model.fit(X_train, y_train)
    xgboost_train_accuracy = model.score(X_train, y_train)
    xgboost_test_accuracy = model.score(X_test, y_test)

    xgboost_feature_importance = importance_fn(model)
    
    return xgboost_feature_importance, xgboost_train_accuracy, xgboost_test_accuracy
```

```python
# Plot feature importances predicted by XGBoost algorithm while plotting it against features.
plot_features("xgboost")   
```

```python
# Scale variables
scaler = StandardScaler()
scaled_train = scaler.fit_transform(X_train)
scaled_test = scaler.transform(X_test)
# Utility functions to scale the target
mean_train = y_train.mean()
std_train = y_train.std()
```

```python
def scale_target(y, mean, std):
    """Use to scale the target variable using the formula => (value - mean)/standard deviation.
    We need to performing this scaling in order to feed it while training the Neural Network"""
    return np.asarray((y - mean)/std)

def reverse_target(pred, mean, std):
     #Use to reverse the scaled target variable using the formula => Predicted_value*standard deviation + mean
     #We need to perform this rescaling in order to compare the predictions"""
    return np.asarray(pred*std + mean)
```

```python
# Set random seed to 42 so that the same set is selected every time
tf.random.set_seed(42)
os.environ['PYTHONHASHSEED'] = str(42)
np.random.seed(42)
random.seed(42)
```

```python
# Build tensorflow session
session_conf = tf.compat.v1.ConfigProto(
    intra_op_parallelism_threads=1,
    inter_op_parallelism_threads=1
)
sess = tf.compat.v1.Session(
    graph=tf.compat.v1.get_default_graph(),
    config=session_conf
)
tf.compat.v1.keras.backend.set_session(sess)
```

```python
# Define NN structure
inp = Input(shape=(scaled_train.shape[1],))
x = Dense(500, activation='relu')(inp)
x = Dense(500, activation='relu')(inp)
x = Dense(500, activation='relu')(inp)
x = Dense(250, activation='relu')(inp)
x = Dense(25, activation='relu')(x)
out = Dense(1)(x)
model = Model(inp, out)
```

```python
# Compile and fit model
model.compile(optimizer='adam', loss='mse', metrics=['mean_squared_error'])
model.fit(scaled_train, scale_target(y_train, mean_train, std_train) , epochs=100, batch_size=64 ,verbose=2)
```

```python
# Compute Permutation and scoring
os.environ['PYTHONHASHSEED'] = str(42)
np.random.seed(42)
random.seed(42)

def dnn():
    """Compute the scores for features"""
    final_score = []
    shuff_pred = []

    for i, col in enumerate(X_test.columns):
        # shuffle column
        shuff_test = scaled_test.copy()
        shuff_test[:, i] = np.random.permutation(shuff_test[:, i])

        # compute score
        score = mean_absolute_error(y_test, reverse_target(model.predict(shuff_test).ravel(), mean_train, std_train))
        shuff_pred.append(reverse_target(model.predict(shuff_test).ravel(), mean_train, std_train))
        final_score.append(score)

    final_score = np.asarray(final_score)
    # Normalize to match scale 0-1.
    final_score = [score*20 for score in final_score]
    
    return final_score
```

```python
# Permutation Importance Report
#  Plot feature importances predicted by Deep Neural Network while plotting it against features.
plot_features("dnn")   
```

```python
#Find average of the feature importances predicted by Random Forest, XGBoost and Neural Network
# while plotting it against features.
plot_features("average")
```

```python
# Compare train and test accuracies to ensure that model is not overfitting on the training set.
_, rf_train_accuracy, rf_test_accuracy = random_forest()
_, xgboost_train_accuracy, xgboost_test_accuracy = xg_boost()

N = 2
rf_accuracies = (rf_train_accuracy, rf_test_accuracy)
print("Random Forest train accuracy: %0.3f" % rf_train_accuracy)
print("Random Forest test accuracy: %0.3f" % rf_test_accuracy)

xgboost_accuracies = (xgboost_train_accuracy, xgboost_test_accuracy)
print("XGBoost train accuracy: %0.3f" % xgboost_train_accuracy)
print("XGBoost test accuracy: %0.3f" % xgboost_test_accuracy)

ind = np.arange(N)
width = 0.1
plt.bar(ind, rf_accuracies, width, label='Random Forest')
plt.bar(ind + width, xgboost_accuracies, width, label='XGBoost')

plt.ylabel('Accuracy')
plt.title('Accuracies by algorithms')

plt.xticks(ind + width / 2, ('Train Accuracy', 'Test Accuracy'))
plt.legend(loc='best')
plt.show()
```

```python
# Prediction error of Deep Neural Network on test set to see if the model is not overfitting
real_pred = reverse_target(model.predict(scaled_test).ravel(), mean_train, std_train)
MAE = mean_absolute_error(y_test, real_pred)
print(MAE)
```
